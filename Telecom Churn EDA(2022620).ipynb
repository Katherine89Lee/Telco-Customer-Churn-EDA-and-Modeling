```python
for col in categorical_columns:
    plt.figure(figsize = (15, 5))
    sns.countplot(telecom[col], color = '#ff355d', order = telecom[col].value_counts().index)
    plt.title('Frequency of '+ col, fontsize=15)
    plt.xticks(rotation=90)
    plt.show();
```

# One Hot Encoding for categorical data that isn't binary

```python
encoded_features = pd.get_dummies(data=telecom, columns=['MultipleLines', 'InternetService', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies', 'Contract', 'PaymentMethod'])
```

```python
pd.concat([telecom, encoded_features],axis=1)
```
